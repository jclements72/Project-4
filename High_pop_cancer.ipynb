```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Read the CSV file  into a Pandas DataFrame
file_path = Path("C:/Users/jordo/OneDrive/Desktop/git-projects/Project-4/Resources/California/California_new_cancer.csv")
file_path2 = Path("C:/Users/jordo/OneDrive/Desktop/git-projects/Project-4/Resources/California/California_cancer_death.csv")
file_path3 = Path("C:/Users/jordo/OneDrive/Desktop/git-projects/Project-4/Resources/Florida/Florida_new_cancer.csv")
file_path4 = Path("C:/Users/jordo/OneDrive/Desktop/git-projects/Project-4/Resources/Florida/Florida_cancer_death.csv")
file_path5 = Path("C:/Users/jordo/OneDrive/Desktop/git-projects/Project-4/Resources/Texas/Texas_new_cancer.csv")
file_path6 = Path("C:/Users/jordo/OneDrive/Desktop/git-projects/Project-4/Resources/Texas/Texas_cancer_death.csv")

Cali_new_df = pd.read_csv(file_path)
Cali_death_df = pd.read_csv(file_path2)

Florida_new_df = pd.read_csv(file_path3)
Florida_death_df = pd.read_csv(file_path4)

Texas_new_df = pd.read_csv(file_path5)
Texas_death_df = pd.read_csv(file_path6)

# Review new cancer DataFrame
Cali_new_df.head()
```

```python
# Review Death DataFrame
Cali_death_df.head()
```

```python
Cali_new_df = Cali_new_df.rename(columns={"Age-Adjusted Rate":"New Case Rate"})
Cali_death_df = Cali_death_df.rename(columns={"Age-Adjusted Rate":"Death Rate"})

Cali_new_df.head()
```

```python
Cali_death_df.head()
```

```python
Florida_new_df = Florida_new_df.rename(columns={"Age-Adjusted Rate":"New Case Rate"})
Florida_death_df = Florida_death_df.rename(columns={"Age-Adjusted Rate":"Death Rate"})

Florida_new_df.head()
```

```python
Florida_death_df.head()
```

```python
Texas_new_df = Texas_new_df.rename(columns={"Age-Adjusted Rate":"New Case Rate"})
Texas_death_df = Texas_death_df.rename(columns={"Age-Adjusted Rate":"Death Rate"})

Texas_new_df.head()
```

```python
Texas_death_df.head()
```

```python
Cali_cancer_df = pd.merge(Cali_new_df, Cali_death_df, how="left", on=["County", "County"])
Cali_cancer_df.head()
```

```python
Cali_cancer_df = Cali_cancer_df[["Area_x", "County", "New Case Rate", "Case Count", "Population_x", "Death Rate", "Death Count"]]
Cali_cancer_df = Cali_cancer_df.rename(columns={"Area_x":"Area", "Population_x": "Population"})
Cali_cancer_df.head()
```

```python
Cali_cancer_df.to_csv('California_cancer', encoding='utf-8')
```

```python
Florida_cancer_df = pd.merge(Florida_new_df, Florida_death_df, how="left", on=["County", "County"])
Florida_cancer_df.head()
```

```python
Florida_cancer_df = Florida_cancer_df[["Area_x", "County", "New Case Rate", "Case Count", "Population_x", "Death Rate", "Death Count"]]
Florida_cancer_df = Florida_cancer_df.rename(columns={"Area_x":"Area", "Population_x": "Population"})
Florida_cancer_df.head()
```

```python
Florida_cancer_df.to_csv('Florida_cancer', encoding='utf-8')
```
