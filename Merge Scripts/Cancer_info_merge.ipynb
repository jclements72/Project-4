```python
#Imports
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame

#New cases
new_cancer_df = pd.read_csv(
    "New York/NY New Cancer Cases.csv")

#Cancer Death cases
cancer_death_df = pd.read_csv(
    "New York/NY Cancer Death.csv")

# Review the DataFrames
new_cancer_df.head()
cancer_death_df.head()
```

```python
#Rename similar columns before merging
new_cancer_df = new_cancer_df.rename(columns={"Age-Adjusted Rate":"Age-Adjusted New Case Rate"})
cancer_death_df = cancer_death_df.rename(columns={"Age-Adjusted Rate":"Age-Adjusted Death Rate"})

#display new dataframe
new_cancer_df.head()
```

```python
#Merge and display new dataframe
school_data_complete = pd.merge(new_cancer_df, cancer_death_df, how="left", on=["County", "County"])
school_data_complete.head()
```

```python
#Select needed columns to new dataframe
cancer_complete = school_data_complete[["Area_x", "County","Age-Adjusted New Case Rate","Case Count","Population_x","Age-Adjusted Death Rate","Death Count"]]

#Display new dataframe
cancer_complete.head()
```

```python
#Renamme columns
state_cancer_df = cancer_complete.rename(columns={"Area_x":"Area", "Population_x": "Population"})
state_cancer_df.head()
```

```python
#Export completed csv
state_cancer_df.to_csv('NY_cancer.csv', encoding='utf-8')
```
