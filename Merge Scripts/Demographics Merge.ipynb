```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
income_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Income_Median_Household_Income_(Both_Sexes_-_2017_to_2021)_2017_-_2021.csv")
income_df.dropna(how='all', axis=1, inplace=True)

uninsured_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Uninsured_Unemployed_(Both_Sexes_-_Ages_16+_-_2017_to_2021)_2017_-_2021.csv")
uninsured_df.dropna(how='all', axis=1, inplace=True)

poverty_white_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Poverty_White_(Both_Sexes_-_2017_to_2021)_2017_-_2021.csv")
poverty_white_df.dropna(how='all', axis=1, inplace=True)

poverty_black_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Poverty_Black_(Both_Sexes_-_2017_to_2021)_2017_-_2021.csv")
poverty_black_df.dropna(how='all', axis=1, inplace=True)

poverty_asian_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Poverty_Asian_Pacific_Islander_(Both_Sexes_-_2017_to_2021)_2017_-_2021.csv")
poverty_asian_df.dropna(how='all', axis=1, inplace=True)

poverty_foreign_born_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Poverty_Foreign_Born_(Both_Sexes_-_2017_to_2021)_2017_-_2021.csv")
poverty_foreign_born_df.dropna(how='all', axis=1, inplace=True)

poverty_native_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Poverty_American_Indian_Alaska_Native_(Both_Sexes_-_2017_to_2021)_2017_-_2021.csv")
poverty_native_df.dropna(how='all', axis=1, inplace=True)

poverty_hispanic_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Poverty_Hispanic_(Both_Sexes_-_2017_to_2021)_2017_-_2021.csv")
poverty_hispanic_df.dropna(how='all', axis=1, inplace=True)

poverty_non_hispanic_df = pd.read_csv(
    "New York/Demographics_New_York_Counties_Poverty_Non-Hispanic_(Origin_Recode_(Both_Sexes_-_2017_to_2021)_2017_-_2021.csv")
poverty_non_hispanic_df.dropna(how='all', axis=1, inplace=True)

poverty_white_df.head()
```

```python
#Ensure all demo dfs have thw same columns and are all uniformly named
poverty_white_df.columns = ["Demographics", "New York Counties", "White Poverty %"]
poverty_black_df.columns = ["Demographics", "New York Counties", "Black Poverty %"]
poverty_asian_df.columns = ["Demographics", "New York Counties", "Asian Poverty %"]
poverty_foreign_born_df.columns = ["Demographics", "New York Counties", "Foreign Poverty %"]
poverty_native_df.columns = ["Demographics", "New York Counties", "Native Poverty %"]
poverty_hispanic_df.columns = ["Demographics", "New York Counties", "Hispanic Poverty %"]
poverty_non_hispanic_df.columns = ["Demographics", "New York Counties", "Non Hispanic Poverty %"]

#Display new df
poverty_asian_df.head()
```

```python
#Display income df before merging
income_df.head()
```

```python
#Create list for multi - dataframe merging
df_list = [income_df,uninsured_df,poverty_white_df,poverty_black_df,poverty_asian_df,poverty_foreign_born_df,poverty_native_df,
            poverty_hispanic_df,poverty_non_hispanic_df,]
```

```python
#import functools for merging
from functools import reduce

#Merge dfs and export to file
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Demographics'],
                                            how='outer'), df_list)
df_merged.to_csv('NY_demo.csv', encoding='utf-8')
#Display dataframe
df_merged.head()
```
