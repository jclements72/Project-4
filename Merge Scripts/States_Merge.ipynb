```python
#Imports
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Import all state dfs
Alaska_df = pd.read_csv(
    "Alaska/Alaska Complete.csv")

Vermont_df = pd.read_csv(
    "Vermont/Vermont_Complete.csv")

Wyoming_df = pd.read_csv(
    "Wyoming/Wyoming_Complete.csv")

California_df = pd.read_csv(
    "California/California_Complete.csv")

Florida_df = pd.read_csv(
    "Florida/Florida_Complete.csv")
Texas_df = pd.read_csv(
    "Texas/Texas_Complete.csv")

#View dfs
Florida_df.head()
```

```python
#Append state dfs together and display new dataframe
state_info_df =  pd.DataFrame().append([Alaska_df,Vermont_df,Wyoming_df,Florida_df,California_df,Texas_df])
state_info_df.head()
```

```python
#View dataframe end due to size
state_info_df.tail()
```

```python
#Remove unneeded columns
state_info_df = state_info_df.drop(columns="Unnamed: 0")
state_info_df.tail()
```

```python
#Export state info df to csv
state_info_df.to_csv('State_Complete_Info.csv', encoding='utf-8')
```
