```python
#Imports
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
#Read in both csvs
demo_df = pd.read_csv(
    "New York/NY_demo.csv")
cancer_df = pd.read_csv(
    "New York/NY_cancer.csv")

#Print dataframe to endure correct importation
cancer_df.head()
```

```python
#View demographics df
demo_df.head()
```

```python
#Remove county from 'county' column for merging
cancer_df['County'] = cancer_df['County'].replace(' County', ' ', regex=True)

cancer_df.head()
```

```python
#Remove state name from 'Demographics' column and rename for merging
demo_df['Demographics'] = demo_df['Demographics'].replace(', New York', ' ', regex=True)

#Rename columns for clearity and uniformity
demo_df = demo_df.rename(columns={'Demographics':'County'})
demo_df = demo_df.rename(columns={'New York Counties_x':'FIPS'})
demo_df = demo_df.rename(columns={'Uninsured':'Uninsured %'})

#display df
demo_df.head()
```

```python
#Merge the two dataframes and display result
complete_df = demo_df.merge(cancer_df, how="left",on="County")
complete_df.head()
```

```python
#Drop unneeded and duplicated columns
complete_df = complete_df.drop(columns=["Unnamed: 0_x", "Unnamed: 0_y"])
#Export complete dataframe to csv
complete_df.to_csv('New_york_Complete.csv', encoding='utf-8')
complete_df
```
