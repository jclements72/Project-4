```python
#Imports
import os
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd)  # Get all the files in that directory
print("Files in %r: %s" % (cwd, files))
```

```python
#Read in California Cancer and Demographics CSVs
Ca_demo_df = pd.read_csv(
    "High_Pop_Resources/California_demo.csv")
Ca_cancer_df = pd.read_csv(
    "High_Pop_Resources/California_cancer.csv")

#Print dataframe to endure correct importation
Ca_cancer_df.head()
```

```python
#Remove county from 'county' column for merging
Ca_cancer_df['County'] = Ca_cancer_df['County'].replace(' County', ' ', regex=True)

Ca_cancer_df.head()
```

```python
Ca_demo_df.head()
```

```python
#Rename FIPS and Income column for merging
Ca_demo_df = Ca_demo_df.rename(columns={'California Counties':'County', 'County FIPS_x': 'FIPS',
                                        'Dollars': 'Income', 'Uninsured Percent': 'Uninsured %'})

#Drop unnecessary FIPS columns
Ca_demo_df = Ca_demo_df.drop(columns=['County FIPS_y', 'County FIPS_x.1', 'County FIPS_y.1', 'County FIPS_x.2',
                                     'County FIPS_y.2', 'County FIPS_x.3', 'County FIPS_y.3'])

#Remove California from County Column
Ca_demo_df['County'] = Ca_demo_df['County'].replace(', California', ' ', regex=True)

Ca_demo_df.head()
```

```python
#Merge the two dataframes and display result
Ca_complete_df = Ca_demo_df.merge(Ca_cancer_df, how="left",on="County")
Ca_complete_df.head()
```

```python
#Export complete dataframe to csv
Ca_complete_df.to_csv('California_Complete.csv', encoding='utf-8')
Ca_complete_df.head()
```
