```python
#Imports
import os
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd)  # Get all the files in that directory
print("Files in %r: %s" % (cwd, files))
```

```python
# Read in California CSVs
Ca_income_df = pd.read_csv("High_Pop_Resources/California/California_Counties_Median_Household_Income_(Both_Sexes_-_2017_to_2021).csv")
Ca_income_df.dropna(how='all', axis=1, inplace=True)

Ca_uninsured_df = pd.read_csv(
    "High_Pop_Resources/California/California_Counties_Uninsured_Unemployed_(Both_Sexes_-_Ages_16+_-_2017_to_2021).csv")
Ca_uninsured_df.dropna(how='all', axis=1, inplace=True)

Ca_poverty_white_df = pd.read_csv(
    "High_Pop_Resources/California/California_Counties_Poverty_White_(Both_Sexes_-_2017_to_2021).csv")
Ca_poverty_white_df.dropna(how='all', axis=1, inplace=True)

Ca_poverty_black_df = pd.read_csv(
    "High_Pop_Resources/California/California_Counties_Poverty_Black_(Both_Sexes_-_2017_to_2021).csv")
Ca_poverty_black_df.dropna(how='all', axis=1, inplace=True)

Ca_poverty_asian_df = pd.read_csv(
    "High_Pop_Resources/California/California_Counties_Poverty_Asian_Pacific_Islander_(Both_Sexes_-_2017_to_2021).csv")
Ca_poverty_asian_df.dropna(how='all', axis=1, inplace=True)

Ca_poverty_foreign_born_df = pd.read_csv(
    "High_Pop_Resources/California/California_Counties_Poverty_Foreign_Born_(Both_Sexes_-_2017_to_2021).csv")
Ca_poverty_foreign_born_df.dropna(how='all', axis=1, inplace=True)

Ca_poverty_native_df = pd.read_csv(
    "High_Pop_Resources/California/California_Counties_Poverty_American_Indian_Alaska_Native_(Both_Sexes_-_2017_to_2021).csv")
Ca_poverty_native_df.dropna(how='all', axis=1, inplace=True)

Ca_poverty_hispanic_df = pd.read_csv(
    "High_Pop_Resources/California/California_Counties_Poverty_Hispanic_(Both_Sexes_-_2017_to_2021).csv")
Ca_poverty_hispanic_df.dropna(how='all', axis=1, inplace=True)

Ca_poverty_non_hispanic_df = pd.read_csv(
    "High_Pop_Resources/California/California_Counties_Poverty_Non-Hispanic_(Origin_Recode_(Both_Sexes_-_2017_to_2021).csv")
Ca_poverty_non_hispanic_df.dropna(how='all', axis=1, inplace=True)
Ca_poverty_hispanic_df.head()
```

```python
# Rename Income and uninsured Area columns
Ca_income_df.columns = ["California Counties", "County FIPS", "Dollars"]
Ca_uninsured_df.columns = ["California Counties", "County FIPS", "Uninsured Percent"]
Ca_income_df.head()
```

```python
#Rename race-specific poverty columns
Ca_poverty_white_df.columns = ["California Counties", "County FIPS", "White Poverty %"]
Ca_poverty_black_df.columns = ["California Counties", "County FIPS", "Black Poverty %"]
Ca_poverty_asian_df.columns = ["California Counties", "County FIPS", "Asian Poverty %"]
Ca_poverty_foreign_born_df.columns = ["California Counties", "County FIPS", "Foreign Poverty %"]
Ca_poverty_native_df.columns = ["California Counties", "County FIPS", "Native Poverty %"]
Ca_poverty_hispanic_df.columns = ["California Counties", "County FIPS", "Hispanic Poverty %"]
Ca_poverty_non_hispanic_df.columns = ["California Counties", "County FIPS", "Non Hispanic Poverty %"]
Ca_poverty_white_df.head()
```

```python
#place all demographic dfs into list for merging
Ca_list = [Ca_income_df,Ca_uninsured_df,Ca_poverty_white_df,Ca_poverty_black_df,Ca_poverty_asian_df,
           Ca_poverty_foreign_born_df,Ca_poverty_native_df,Ca_poverty_hispanic_df,Ca_poverty_non_hispanic_df,]
```

```python
#Merge list of dfs into one dataframe
from functools import reduce

Ca_merged = reduce(lambda  left,right: pd.merge(left,right,on=["California Counties"],
                                            how='outer'), Ca_list)

Ca_merged.to_csv('California_demo.csv', encoding='utf-8')

Ca_merged.head()
```
