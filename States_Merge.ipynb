```python
#Imports
import os
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd)  # Get all the files in that directory
print("Files in %r: %s" % (cwd, files))
```

```python
#Load in all states csvs
Alaska_df = pd.read_csv(
    "Resources/Alaska Complete.csv")

Vermont_df = pd.read_csv(
    "Resources/Vermont_Complete.csv")

Wyoming_df = pd.read_csv(
    "Resources/Wyoming_Complete.csv")

California_df = pd.read_csv(
    "Resources/California_Complete.csv")

Florida_df = pd.read_csv(
    "Resources/Florida_Complete.csv")

Texas_df = pd.read_csv(
    "Resources/Texas_Complete.csv")

Wyoming_df.head()
```

```python
#Append all state dfs into one dataframe
state_info_df =  pd.DataFrame().append([Alaska_df,Vermont_df,Wyoming_df,Florida_df,California_df,Texas_df])
state_info_df.head()
```

```python
#Ensure all states are in df
state_info_df.tail()
```

```python
#Drop unneeded column
state_info_df = state_info_df.drop(columns="Unnamed: 0")
state_info_df.head()
```

```python
#Export dataframe to csv
state_info_df.to_csv('State_Complete_Info.csv', encoding='utf-8')
```
