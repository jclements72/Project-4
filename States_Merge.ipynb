```python
#Imports
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
#Load in all states csvs
Alaska_df = pd.read_csv(
    "Alaska/Alaska Complete.csv")

Vermont_df = pd.read_csv(
    "Vermont/Vermont_Complete.csv")

Wyoming_df = pd.read_csv(
    "Wyoming/Wyoming_Complete.csv")

California_df = pd.read_csv(
    "California/California_Complete")

Florida_df = pd.read_csv(
    "Florida/Florida_Complete")

Texas_df = pd.read_csv(
    "Texas/Texas_Complete")

Wyoming_df.head()
```

```python
#Append all state dfs into one dataframe
state_info_df =  pd.DataFrame().append([Alaska_df,Vermont_df,Wyoming_df,Florida_df,California_df,Texas_df])
state_info_df.head()
```

```python
#Ensure all states are in df
state_info_df.tail()
```

```python
#Drop unneeded column
state_info_df = state_info_df.drop(columns="Unnamed: 0")
state_info_df.head()
```

```python
#Export dataframe to csv
state_info_df.to_csv('State_Complete_Info', encoding='utf-8')
```
