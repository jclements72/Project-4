```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
new_cancer_df = pd.read_csv(
    "wyomin_new_can.csv")

cancer_death_df = pd.read_csv(
    "wyomin_cancer_death.csv")
# Review the DataFrame
new_cancer_df.head()
cancer_death_df.head()
```

```python
new_cancer_df = new_cancer_df.rename(columns={"Age-Adjusted Rate":"Age-Adjusted New Case Rate"})
cancer_death_df = cancer_death_df.rename(columns={"Age-Adjusted Rate":"Age-Adjusted Death Rate"})

new_cancer_df.head()
```

```python
school_data_complete = pd.merge(new_cancer_df, cancer_death_df, how="left", on=["County", "County"])
school_data_complete.head()
```

```python
school_data_complete = school_data_complete[["Area_x", "County","Age-Adjusted New Case Rate","Case Count","Population_x","Age-Adjusted Death Rate","Death Count"]]

school_data_complete.head()
```

```python
alaska_cancer_df = school_data_complete.rename(columns={"Area_x":"Area", "Population_x": "Population"})
alaska_cancer_df.head()
```

```python
alaska_cancer_df.to_csv('wyoming_cancer', encoding='utf-8')
```
